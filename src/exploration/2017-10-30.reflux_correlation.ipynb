# Within-patient prevalence of exchanged bugs

Define within-patient prevalence of exchanged bugs as the percent of bugs each patient has in both their lungs and stomach.

**Pseudo-code**

1. For each site combination,  
    1. Get the exchanged OTUs
    1. Get the patients with those two sites, and the associated samples
    1. For each patient (prob list comprehension), 
        1. Calculate how many of the OTUs are present in both sites.
    1. Store the site combination, patient, samples, and patient metadata

```python
fnmeta = '/Users/claire/github/aspiration-analysis/data/clean/rosen.metadata.clean'
fnotu = '../../data/clean/rosen.otu_table.rel_abun.clean'
fnexchanged = '../../data/analysis/exchange.labeled_otus.txt'

meta = pd.read_csv(fnmeta, sep='\t', index_col=0)
df = pd.read_csv(fnotu, sep='\t', index_col=0)
exchanged = pd.read_csv(fnexchanged, sep='\t', index_col=0)

# Make tidy OTU table with 'site' and 'subject_id' columns
df.index.name = 'sample'
dflong = pd.melt(df.reset_index(), id_vars=['sample'], var_name='otu',
                 value_name='abun')
dflong = dflong.merge(
            meta[['site', 'subject_id']],
            left_on='sample', right_index=True)
dflong.head()
```

```python
sites = ['bal', 'gastric_fluid', 'throat_swab']
lstpreva = []

for site1 in sites:
    for site2 in sites[sites.index(site1)+1:]:
        sitecomp = site1 + '-' + site2

        # Get patients with the two sites
        patients = dflong\
                    .query('(site == @site1) | (site == @site2)')\
                    .groupby(['subject_id', 'site'])\
                    .size()\
                    .reset_index()\
                    .groupby('subject_id')\
                    .size()
        patients = patients[patients == 2].index.tolist()

        # Get the exchanged OTUs btw the two sites
        otus = exchanged[sitecomp].dropna().index.tolist()
        print(sitecomp, len(otus))
        
        # Get the abundances of the shared otus in the patients with the two sites
        tmp1 = dflong.query('otu == @otus').query('subject_id == @patients').query('site == @site1')
        tmp2 = dflong.query('otu == @otus').query('subject_id == @patients').query('site == @site2')

        # Merge them, to have df with [patient, otu, abun in site1, abun in site2]
        sitesdf = pd.merge(tmp1, tmp2, 
            left_on=['subject_id', 'otu'], 
            right_on=['subject_id', 'otu'],
            suffixes=('_site1', '_site2'))
        
        # Calculate number of exchanged OTUs each patient has in both sites
        for p, pt_df in sitesdf.groupby('subject_id'):
            # Count number of sites that each OTU is non-zero in
            numpresent = (pt_df[['abun_site1', 'abun_site2']] > 0).sum(axis=1)
            preva = sum(numpresent == 2)/float(len(otus))
            
            lstpreva.append([p, site1, site2, sitecomp, len(otus), preva])
```

```python
dfpreva = pd.DataFrame(lstpreva, 
    columns=['subject_id', 'site1', 'site2', 'site_comparison', 'n_shared', 'pt_prevalence'])
dfpreva.head()
```

```python
# Add aspiration and reflux metadata to these values
refluxcols = ['number of full colum events/total events',
              'Total number of reflux episodes (acid+non-acid)',
              'percent proximal total',
              'percent distal total']

dfpreva = pd.merge(dfpreva, meta[['subject_id', 'mbs_consolidated'] + refluxcols], 
                   left_on='subject_id', right_on='subject_id',
                   how='left')
```

## Aspiration vs. within-patient prevalence

Biologically, I'm slightly more interested in the OTU-centric version, because this can also identify OTUs which are way way more often present in both sites of aspirators vs. non-aspirators (meaning they're being preferentially aspirated, maybe). ## Reflux correlations

Okay, now the real point of this stuff: does a patient's reflux correlate with how often the exchange bugs are in both of their sites?

```python
dfpreva.head()
```

```python
balgas = dfpreva.query('site_comparison == "bal-gastric_fluid"')
balgas = balgas.replace('n/a', np.nan)
balgas['number of full colum events/total events'] = balgas['number of full colum events/total events'].replace("1.0", np.nan)

toplot = pd.melt(balgas, value_vars=refluxcols, value_name='reflux_value',
                 var_name='reflux_type',
                 id_vars=['pt_prevalence', 'mbs_consolidated'])
toplot['reflux_value'] = toplot['reflux_value'].astype(float)
toplot.head()
```

```python
sns.set_style('white')
colors = {'Normal': 'green', 'Aspiration/Penetration': 'blue', 'nan': 'gray'}

toplot['mbs_consolidated'] = toplot['mbs_consolidated'].replace(np.nan, 'nan')

g = sns.FacetGrid(
    data=toplot, 
    col='reflux_type', col_wrap=2, hue='mbs_consolidated', 
    sharex=False, legend_out=True, palette=colors,
    aspect=1.5)
g.map(plt.scatter, 'reflux_value', 'pt_prevalence').add_legend().set_titles("{col_name}")
```

```python
# Calculate correlation for funsies
from scipy.stats import spearmanr, pearsonr

for r, subdf in toplot.groupby('reflux_type'):
    print(r, spearmanr(subdf.dropna()['reflux_value'], subdf.dropna()['pt_prevalence']))
```
