## Data

This first part is just me preparing the data for the later analyses. You can skip down to the "Analysis" section.

What I do in this section:

2. Clean up the data a bit
3. Define which columns I'm looking at for bile and reflux metadata

```python
fnjsd = '/Users/claire/github/aspiration-analysis/data/analysis/jsd.txt'

jsd = pd.read_csv(fnjsd, sep='\t').dropna(how='all', axis=1)
jsd.head()
```

```python
# These are all the reflux-related columns in the full metadata
refluxcols = ['total duration of acid reflux',
              'Reflux - total number of episodes',
              'Number of acid reflux episodes',
              'Total number of reflux episodes (acid+non-acid)',
              'SI - Reflux', 'SSI - Reflux', 'SAP - Reflux',
              'Number of non-acid reflux episodes',
              'percent distal nonacid', 'percent proximal total',
              'percent distal acid', 'percent proximal acid',
              'percent proximal nonacid', 'percent distal total',
              'number of full colum events/total events',
              'Number of full column episodes']
refluxcols = [i for i in refluxcols if i in jsd.columns]
refluxcols
```

```python
bilecols = ['Was Bile CA detected?',
            'Was Bile DCA detected?',
            'Was Bile LCA detected?',
            'Was Bile TCA detected?']
bilecols = [i for i in bilecols if i in jsd.columns]
bilecols
```

```python
# Do some data cleaning.
jsd[bilecols] = jsd[bilecols].replace('NQ Not quantifiable', 'No')
jsd[bilecols] = jsd[bilecols].replace('ND Not detectable', 'No')

jsd[refluxcols] = jsd[refluxcols].replace('n/a', np.nan)
jsd[refluxcols] = jsd[refluxcols].astype(float)
jsd['number of full colum events/total events'] = jsd['number of full colum events/total events'].replace(1, np.nan)
```

```python
# Manipulate dataframe so that it can be faceted bile vs. reflux
tidyjsd = pd.melt(jsd, value_vars=bilecols, 
                  var_name='bile_type', value_name='bile_value',
                  id_vars=[i for i in jsd.columns if i not in bilecols])
tidyjsd['bile_type'] = tidyjsd['bile_type']\
    .apply(lambda x: x.split('Was Bile')[1].split('detected?')[0].strip())

tidyjsd = pd.melt(tidyjsd, value_vars=refluxcols, 
                  var_name='reflux_type', value_name='reflux_value',
                  id_vars=[i for i in tidyjsd.columns if i not in refluxcols])
tidyjsd.dropna(how='all').head()
```

```python
# Look at how many patients have data for each of the reflux measures
for r, subdf in tidyjsd.groupby('reflux_type'):
    print(r, subdf.loc[subdf['subject'].drop_duplicates().index].dropna(subset=['reflux_value']).shape[0])
```

```python
## Looks like Reflux - total number, and the SAP, SI, and SSI measures aren't very filled out. Remove them.
remove_reflux_cols = ['Reflux - total number of episodes',
                     'SAP - Reflux',
                     'SI - Reflux',
                     'SSI - Reflux']
tidyjsd = tidyjsd[~tidyjsd['reflux_type'].isin(remove_reflux_cols)]
```

### Aspiration and bile acid in the lungs

*Are aspirators more likely than non-aspirators to have bile acid detected in their lungs?*

These results make me think that perhaps the more relevant question is simply whether or not aspirators are more likely than non-aspirators to have bile acids in their lungs. Let's test this with a Fisher's test.

The following code essentially prints out confusion tables showing how many of the Normal/Aspirators have/don't have the bile acid. The p-value printed is the results from a Fisher's exact test on this confusion matrix.


```python
fisherdf = tidyjsd[['subject', 'bile_type', 'bile_value', 'mbs_consolidated']].dropna().drop_duplicates()
for bile, subdf in fisherdf.groupby('bile_type'):
    print('Bile acid: {} '.format(bile))
    print(pd.crosstab(subdf['bile_value'], subdf['mbs_consolidated']))
    print('p = {:.3f}'.format(fisher_exact(pd.crosstab(subdf['bile_value'], subdf['mbs_consolidated']).values)[1]))
    print('\n')
```

**Inconclusive**. There aren't enough aspirators who also have the bile acid testing to be able to say anything here.

### Reflux and community similarities

*Do patients with higher reflux have more similar gastric-lung communities?*

Let's leave the bile acid data behind, and go back to the 16S communities.

If we hypothesize that reflux increases the connection between lungs and stomach, then we would expect to see more similar lung and gastric communities in patients with more reflux. Moreover, if aspirators are less able to protect their lungs from gastric exchange, we would expect such a relationship to be even more pronounced in aspirators than non-aspirators.

#### How to read the following plots:

* Each plot shows a different type of reflux measurement, labeled above each plot.
* The x-axis is the value for the reflux measurement.
* The y-axis is the dissimilarity between the gastric and lung communities. 1 means completely different communities; 0 means completely identical communities.
* Points are colored by aspiration status: green for Normal, blue for Aspirators, and gray for patients without this metadata

I've also printed the Spearman rank correlation and associated p-values for this data below the plot. I calculated the correlation across all patients as well as just within Normal patients and just within Aspirators.

```python
# Plot beta diversity between the three sites vs. reflux value for each reflux col
colors = {'Normal': 'green', 'Aspiration/Penetration': 'blue', 'nan': 'gray'}
tidyjsd['mbs_consolidated'] = tidyjsd['mbs_consolidated'].replace(np.nan, 'nan')
g = sns.FacetGrid(
    data=tidyjsd.query('site_comparison == "bal_gastric"').query('patient_comp == "within"'), 
    col='reflux_type', col_wrap=4, hue='mbs_consolidated', sharex=False, legend_out=True,
    palette=colors)
g.map(plt.scatter, 'reflux_value', 'beta').add_legend()
```

```python
# Same plot as above, but only for the "total" reflux cols
sns.set_style('white')
colors = {'Normal': 'green', 'Aspiration/Penetration': 'blue', 'nan': 'gray'}
tidyjsd['mbs_consolidated'] = tidyjsd['mbs_consolidated'].replace(np.nan, 'nan')

refluxkeep = ['number of full colum events/total events',
              'Total number of reflux episodes (acid+non-acid)',
              'percent proximal total',
              'percent distal total']

toplot = tidyjsd\
            .query('site_comparison == "bal_gastric"')\
            .query('patient_comp == "within"')\
            .query('reflux_type == @refluxkeep')
            
g = sns.FacetGrid(
    data=toplot, 
    col='reflux_type', col_wrap=2, hue='mbs_consolidated', 
    sharex=False, legend_out=True, palette=colors,
    aspect=1.5)
g.map(plt.scatter, 'reflux_value', 'beta').add_legend().set_titles("{col_name}")
```

```python
# Print the spearman correlation between gastric-bal JSD and reflux,
# 1. over all patients
# 2. just in normal patients
# 3. just in aspirators
for_corr = tidyjsd[['sample1', 'sample2', 'subject', 
                    'patient_comp', 'site_comparison', 
                    'reflux_type', 'reflux_value', 
                    'mbs_consolidated', 'jsd']]\
           .drop_duplicates()

for_corr['mbs_consolidated'] = for_corr['mbs_consolidated'].replace('nan', np.nan)
for_corr = for_corr.query('site_comparison == "bal_gastric"').query('patient_comp == "within"')

corrs_lst = []
for g, subdf in for_corr.groupby('reflux_type'):
    print(g)
    subdf = subdf.dropna(subset=['reflux_value'])
    rall = spearmanr(subdf['reflux_value'], subdf['jsd'])
    corrs_lst.append([g, 'All patients', subdf.shape[0], rall[0], rall[1]])
    print('\tAll patients   : r = {:.3f}, p = {:.3f}'.format(rall[0], rall[1]))
    
    aspdf = subdf.query('mbs_consolidated == "Aspiration/Penetration"')
    rasp = spearmanr(aspdf['reflux_value'], aspdf['jsd'])
    corrs_lst.append([g, 'Aspirators', aspdf.shape[0], rasp[0], rasp[1]])
    print('\tAspirators ({}) : r = {:.3f}, p = {:.3f}'.format(aspdf.shape[0], rasp[0], rasp[1]))

    nmldf = subdf.query('mbs_consolidated == "Normal"')
    rnml = spearmanr(nmldf['reflux_value'], nmldf['jsd'])
    corrs_lst.append([g, 'Normal', nmldf.shape[0], rnml[0], rnml[1]])
    print('\tNormal  ({})   : r = {:.3f}, p = {:.3f}\n'.format(nmldf.shape[0], rnml[0], rnml[1]))
```

We're looking for negative correlations: as the reflux severity increases (x-axis), the communities get more similar (beta-diversity, y-axis, decreases).

### Results:

1. None of the reflux measurements are correlated with gastric/bal similarity across all patients, except perhaps `number of full column events/total number of events`.
2. `Number of full column events` is also the only reflux measurement that seems correlated with bal/gastric community similarity in `Normal` patients.
3. In `Aspirators`, `percent distal acid`, `percent distal total`, and `percent proximal acid` seem negatively correlated with the dissimilarity. In other words, the more distal acid, proximal acid, or distal total reflux that an aspirator patient has, the more similar their lung and gastric communities are.
4. However, this is still incredibly preliminary: we have very small sample sizes (only 8 aspirators with reflux testing!)

It's hard to reconcile these results with your previous work (which found that patients with positive BAL culture have higher proximal non-acid reflux). To me, these results indicate that there *may* be a relationship in aspirators, but that we need larger sample sizes.
