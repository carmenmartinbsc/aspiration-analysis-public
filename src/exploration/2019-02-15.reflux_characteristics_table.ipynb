```python
import pandas as pd
import numpy as np
```

```python
freflux = '../../data/raw/metadata/reviewer-response-jan2019/k23IMPEDANCERESULTS.csv'
fpts = '../../final/supp_files/patients_used_in_each_analysis.csv'

meta = pd.read_csv(freflux, sep=',')
meta.head()
```

```python
df_pts = pd.read_csv(fpts, sep=',', index_col=0)
df_pts.head()
```

```python
reflux_pts = df_pts[ df_pts['figure5.reflux_correlation'] == True ].index.tolist()
len(reflux_pts)
```

```python
refluxmeta = meta.query('SUBJID == @reflux_pts')
refluxmeta.shape
```

```python
missing_subjs = [i for i in reflux_pts if i not in refluxmeta['SUBJID'].tolist()]
missing_subjs
```
