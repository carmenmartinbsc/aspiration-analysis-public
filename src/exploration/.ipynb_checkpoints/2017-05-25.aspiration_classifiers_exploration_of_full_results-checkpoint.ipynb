The goal of this notebook is to explore and visualize the different classifiers made by `make_classifiers.py`.

This notebook is different than the other `aspiration_classifier` notebook because it just reads in the pre-made classifiers' data and pokes around those (rather than building classifiers itself, as in the older notebook).


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

# to pretty print dataframes
from IPython.display import display

sns.set_style('white')
```

```python
# Raw data
fnotu = '/Users/claire/github/aspiration-analysis/data/clean/rosen.otu_table.rel_abun.clean'
fnmeta = '/Users/claire/github/aspiration-analysis/data/clean/rosen.metadata.clean'

df = pd.read_csv(fnotu, sep='\t', index_col=0)
meta = pd.read_csv(fnmeta, sep='\t', index_col=0)

# Classifiers
fnclssum = '/Users/claire/github/aspiration-analysis/data/analysis/rf_results.summary_stats.txt'
fnclsroc = '/Users/claire/github/aspiration-analysis/data/analysis/rf_results.rocs.txt'
fnclsfeats = '/Users/claire/github/aspiration-analysis/data/analysis/rf_results.features.txt'

summary = pd.read_csv(fnclssum, sep='\t')
rocs = pd.read_csv(fnclsroc, sep='\t')
feats = pd.read_csv(fnclsfeats, sep='\t')

summary['log_fisher_p'] = np.log10(summary['fisher_p'])
```

### Tidy OTU


```python
tidyotu = pd.melt(df.reset_index(), id_vars='index', value_name='abun', var_name='otu')
tidyotu = tidyotu.rename(columns={'index': 'sample'})
keepcols = ['mbs_consolidated', 'site', 'subject_id']

tidyotu = tidyotu.merge(meta[keepcols], left_on='sample', right_index=True)
tidyotu['otu_w_site'] = tidyotu['otu'] + '--' + tidyotu['site']
tidyotu['log_abun'] = np.log10(tidyotu['abun'].replace(0, 1e-6))
```

```python
fig, ax = plt.subplots(2, 1)
sns.boxplot(x='site', y='auc', data=summary, ax=ax[0], hue='shuffled')
sns.stripplot(x='site', y='auc', data=summary, linewidth=0.6, jitter=True, 
              ax=ax[0], hue='shuffled', split=True)
ax[0].set_xticklabels('')
ax[0].set_xlabel('')
ax[0].set_title("AUC")
ax[0].legend_.remove()

sns.boxplot(x='site', y='log_fisher_p', data=summary, ax=ax[1], hue='shuffled')
sns.stripplot(x='site', y='log_fisher_p', data=summary, linewidth=0.6, jitter=True, 
              ax=ax[1], hue='shuffled', split=True)
_ = ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90)
ax[1].set_title('fisher P value')
ax[1].legend_.remove()
```

```python
fig, ax = plt.subplots(3, 3, figsize=(10, 10))
ax = ax.flatten()
i = 0

groups = ['bal', 'throat_swab', 'gastric_fluid',
          'bal_and_throat_swab', 'gastric_fluid_and_throat_swab', 
          'bal_and_gastric_fluid', 'all_sites']
for g in groups:
    subdf = rocs.query('site == @g')
    for shuffled, color in zip([True, False], ['gray', 'b']):
        tmp = subdf.query('shuffled == @shuffled')
        for n, onerep in tmp.groupby('replicate'):
            ax[i].plot(onerep['mean_fpr'], onerep['mean_tpr'], color, alpha=0.5)
    ax[i].set_title(g)
    ax[i].set_xticklabels('')
    ax[i].set_yticklabels('')
    i += 1

ax[7].set_axis_off()
ax[8].set_axis_off()
```

```python
fig, ax = plt.subplots(3, 3, figsize=(10, 10))
ax = ax.flatten()
i = 0

groups = ['bal', 'throat_swab', 'gastric_fluid',
          'bal_and_throat_swab', 'gastric_fluid_and_throat_swab', 
          'bal_and_gastric_fluid', 'all_sites']

for g in groups:
    subdf = rocs.query('site == @g')
    meanx = subdf.groupby(['shuffled', 'mean_fpr']).mean()
    stdx = subdf.groupby(['shuffled', 'mean_fpr']).std()
    
    for shuffled, color in zip([True, False], ['gray', 'b']):
        x = meanx.loc[shuffled].index
        mu = meanx.loc[shuffled]['mean_tpr'].values
        sigma = stdx.loc[shuffled]['mean_tpr'].values
        ax[i].fill_between(x, mu+sigma, mu-sigma, color=color, alpha=0.2)
        ax[i].plot(x, meanx.loc[shuffled, 'mean_tpr'].values, color)
    
    ax[i].set_xlim([0, 1])
    ax[i].set_ylim([0, 1])
    ax[i].set_title(g)
    ax[i].set_xticklabels('')
    ax[i].set_yticklabels('')
    
    a, p = summary\
            .query('site == @g')\
            .query('shuffled == False')\
            [['auc', 'fisher_p']]\
            .mean()
    a = 'AUC = {:.2f}'.format(a)
    p = 'p = {:.3f}'.format(p)
    ax[i].text(0.65, 0.2, a)
    ax[i].text(0.65, 0.1, p)
    i += 1

ax[7].set_axis_off()
ax[8].set_axis_off()
```

Current thinking: throat swab alone is decent-ish classifiers, but not great. Gastric and throat have some potential, but looking at ROC curves seems less good.

If you combine throat and BAL together, it improves on the individual classifiers. If you add gastric to BAL, it makes it worse. If you add gastric to throat, you basically recover the same performance as throat alone. So there's something about the relationship between throat and BAL that's going on in the aspirators vs. non-aspirators.

## Features

Let's look at the features, see if anything comes out here...

```python
feats.head()
```

Each replicate here is a re-run of the classifier. I made each classifier 25 times on the entire training data. These are the features that came out as important.

```python
N_feats = 30

fig, ax = plt.subplots(2, 4, figsize=(10,4))
ax = ax.flatten()
i = 0

groups = ['bal', 'throat_swab', 'gastric_fluid', 'bal_and_gastric_fluid',
          'bal_and_throat_swab', 'gastric_fluid_and_throat_swab', 'all_sites']

for g in groups:
    subdf = feats.query('site == @g').query('shuffled == False')
    # Normalize feature importances within replicates
    tmp = subdf.groupby('replicate')['importance'].transform(lambda x: x/x.max()).rename('norm_imp')
    subdf = subdf.join(tmp)
    feat_order = subdf.groupby('feature').agg('mean').sort_values(by='norm_imp', ascending=False).index
    sns.boxplot(data=subdf, x='feature', y='norm_imp', order=feat_order[:N_feats], ax=ax[i])
    ax[i].set_xticklabels('')
    ax[i].set_xlabel('')
    ax[i].set_title(g)
    i += 1
```

In all cases (except the all_sites classifier), it looks like the top feature is quite interesting. Let's pull it out and see how often it's the same one...

Or perhaps I can color these by whether they're BAL, gastric, or throat OTUs??

```python
for g in groups:
    subdf = feats.query('site == @g').query('shuffled == False')
    # Normalize feature importances within replicates
    tmp = subdf.groupby('replicate')['importance'].transform(lambda x: x/x.max()).rename('norm_imp')
    subdf = subdf.join(tmp)
    feat_order = subdf.groupby('feature').agg('mean').sort_values(by='norm_imp', ascending=False).index
    print(g, feat_order[0])
```

Interesting... The top feature for the `bal`, `throat_swab`, and `bal_and_throat_swab` classifiers are all different. I wonder if this is because of slight differences in the ranking of features, or because the three different classifiers are looking for different things?
```python
# Get the top features from each classifier, put it in a dataframe
allfeats = {}
for g in groups:
    subdf = feats.query('site == @g').query('shuffled == False')
    # Normalize feature importances within replicates
    tmp = subdf.groupby('replicate')['importance'].transform(lambda x: x/x.max()).rename('norm_imp')
    subdf = subdf.join(tmp)
    feat_order = subdf.groupby('feature').agg('mean').sort_values(by='norm_imp', ascending=False).index
    allfeats[g] = feat_order
```

Look at the abundance of top five features in the 3 classifiers in aspirators vs. non-aspirators

```python
keepgroups = ['bal', 'throat_swab', 'bal_and_throat_swab']
keepfeats = list(set([i for k in keepgroups for i in allfeats[k][0:5]]))
len(keepfeats)
```

```python
fig, ax = plt.subplots(3, 5, figsize=(10,5))
ax = ax.flatten()
i = 0

for g in keepgroups:
    for f in allfeats[g][0:5]:
        sns.stripplot(data=tidyotu.query('otu_w_site == @f'), 
                      x='mbs_consolidated', y='log_abun', 
                      jitter=True, linewidth=0.5, ax=ax[i])
        if i <= 9:
            ax[i].set_xticklabels('')
            ax[i].set_xlabel('')
        if i not in [0, 5, 10]:
            ax[i].set_yticklabels('')
            ax[i].set_ylabel('')
        if i in [4, 9, 14]:
            ax[i].set_ylabel(g)
            ax[i].yaxis.set_label_position("right")
        i += 1
        
fig.suptitle('top 5 features in bal, throat_swab, and bal+throat_swab classifiers')
```
