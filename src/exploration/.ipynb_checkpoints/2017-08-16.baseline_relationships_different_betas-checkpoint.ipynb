This script continues looking at how to show the baseline relationship between the sites.

Specifically, I'm hoping to:
- Make a PCA/PCoA showing that aerodigestive sites overlap much more than rectal/stool swabs
- Compare the intra-patient beta diversities for site comparisons (including unifrac as well as JSD)
- Show that throat-lung comparison is always the least similar across all patients

# Show lung-throat are least similar within patients


But first - related question motivating this: connect the dots between patients on these plots above...


```python
xmap = {'bal_throat': 0, 'bal_gastric': 1, 'gastric_throat': 2}

fig, ax = plt.subplots(1, 2)

for subj, subdf in jsd.query('site_comparison == @site_comparisons').groupby('subject'):
    if 'bal_throat' in subdf['site_comparison'].values:
        if 'bal_gastric' in subdf['site_comparison'].values:
            ax[0].plot([0, 1], 
                [subdf.query('site_comparison == "bal_throat"')['beta'].values[0], 
                 subdf.query('site_comparison == "bal_gastric"')['beta'].values[0]])
        if 'gastric_throat' in subdf['site_comparison'].values:
            ax[1].plot([0, 1], 
                [subdf.query('site_comparison == "bal_throat"')['beta'].values[0], 
                 subdf.query('site_comparison == "gastric_throat"')['beta'].values[0]])
```

```python
# Okay let's do some stats on this
dup_subjs = ['01-263-4', '01-230-9'] # These have two gastric_throat comparisons
difjsd = jsd.query('site_comparison == @site_comparisons')\
            .query('subject != @dup_subjs')\
            .pivot(index='subject', columns='site_comparison', values='beta')
difjsd['bal_thr-bal_gas'] = difjsd['bal_throat'] - difjsd['bal_gastric']
difjsd['bal_thr-gas_thr'] = difjsd['bal_throat'] - difjsd['gastric_throat']
difjsd['gas_thr-bal_gas'] = difjsd['gastric_throat'] - difjsd['bal_gastric']
```

```python
difjsd.dropna(how='all', axis=1).head(10)
```

```python
## This plot is confusing. It's the difference in JSD between different site comparisons

#tmp = pd.melt(difjsd.reset_index(), value_vars=['bal_thr-bal_gas', 'bal_thr-gas_thr', 'gas_thr-bal_gas'])
#sns.boxplot(data=tmp, x='site_comparison', y='value', fliersize=0)
#sns.stripplot(data=tmp, x='site_comparison', y='value', split=True, jitter=True, size=5, linewidth=0.6)
```

```python
# These t-tests are testing whether two sites are significantly more similar than two other sites.
# In others words, are the lungs and throat significantly more similar than the lungs and stomach?

# This is just the test between my intra-patient JSD boxplots, but with paired samples.

from scipy.stats import ttest_rel

def get_ttest_val(comp1, comp2, df):
    """
    Calculate paired t-test between comp1 and comp2 columns in df.
    """    
    tmp = difjsd[[comp1, comp2]].dropna()
    return ttest_rel(tmp[comp1], tmp[comp2])

for c1 in site_comparisons:
    for c2 in site_comparisons[site_comparisons.index(c1)+1:]:
        print(c1, c2, get_ttest_val(c1, c2, difjsd))
```

>a.	Sample IDs are indicated in the “SampleID” column as follows:

2.	Initial stool sample indicated by patient ID followed by “SI”

3.	Final throat swab indicated by patient ID followed by “TF”

4.	Final stool sample indicated by patient ID followed by “SF”

So these throat-stool comparisons are all intra-patient, but sometimes correspond to both samples being taken at the same time, whereas other times correspond to samples taken at different times. tbh, I don't think this matters that much - but just to be safe, let's try to grab just the same-time-point comparisons.

```python
# Grab the stool_throat samples that correspond to the same time point
tmp = jsd.query('site_comparison == "stool_throat"')\
         .groupby('subject').size()
subjs = list(tmp[tmp > 1].index)
keep = jsd.query('subject == @subjs')
keep = keep[[i[-1] == j[-1] for i, j in zip(keep['sample1'], keep['sample2'])]]
keep.head()
```

```python
neworder = ['bal_throat', 'bal_gastric', 'gastric_throat', 
            'rectalswab_throat', 
            'gastric_rectalswab', 'stool_stool']

# This newjsd has only the same time point comparisons for stool_throat samples. 
# Not sure if any of the others similarly need correction for time point...
newjsd = pd.concat((jsd.query('site_comparison == @neworder'), jsd.loc[keep.index]))

neworder = ['bal_throat', 'bal_gastric', 'gastric_throat', 
            'stool_throat', 'rectalswab_throat', 
            'gastric_rectalswab', 'stool_stool']

fig = plt.figure(figsize=(10,4))
sns.boxplot(data=newjsd, x='site_comparison', y='beta', fliersize=0, order=neworder)
sns.stripplot(data=newjsd, x='site_comparison', y='beta', order=neworder,
              split=True, jitter=True, size=5, linewidth=0.6)
plt.title('intra-patient comparisons')
```

```python
# Statistical sig between these comparisons
from scipy.stats import ranksums
from statsmodels.sandbox.stats.multicomp import multipletests

pvals = []

for c1 in neworder:
    for c2 in neworder[neworder.index(c1)+1:]:
        res = ranksums(
                newjsd.query('site_comparison == @c1')['beta'], 
                newjsd.query('site_comparison == @c2')['beta'])
        pvals.append([c1, c2, res[0], res[1]])
pvals = pd.DataFrame(pvals, columns=['comp1', 'comp2', 'stat', 'p'])
pvals['q'] = multipletests(pvals['p'])[1]
pvals.sort_values(by='q')
```
