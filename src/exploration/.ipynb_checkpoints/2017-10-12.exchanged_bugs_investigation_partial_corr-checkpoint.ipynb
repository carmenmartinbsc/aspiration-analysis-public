This notebook explores the new "exchanged" bacteria, calculated using the partial correlation vs. the old way (without considering the third site)

````python
import pandas as pd
import numpy as np
from statsmodels.sandbox.stats.multicomp import multipletests

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from IPython.display import display
````

````python
fname = '../../data/analysis/exchange.with_partial_corrs.txt'
df = pd.read_csv(fname, sep='\t')
````

````python
df['site_comparison'] = df['site1'] + '-' + df['site2']
df.head()
````

````python
df['site_comparison'].unique()
````

Let's compare the number of bacteria with positive correlations with and without controlling for the third sites.

````python
# Number of patients necessary for comparison
# Note: the n_site12 is patients with non-zero abundance in sites1 and 2
# The n_partial is the patients in n_site12 which had the third site sequenced,
# and so will usually be lower than n_site12. For now, let's condition on
# n_partial.
n_thresh = 10

# Correct for multiple hypothesis testing after excluding OTUs without enough patients
df = df.query('n_partial >= @n_thresh')
_, df['q_site12'], _, _ = multipletests(df['p_site12'], method='fdr_bh')
_, df['q_partial'], _, _ = multipletests(df['p_partial'], method='fdr_bh')
df.head()
````

````python
display(df.query('n_partial >= @n_thresh').query('r_site12 > 0').query('q_site12 < 0.1').groupby('site_comparison').size())
#display(df.query('n_partial >= @n_thresh').query('r_site12 > 0').groupby('site_comparison').size())

display(df.query('n_partial >= @n_thresh').query('r_partial > 0').query('q_partial < 0.1').groupby('site_comparison').size())
#display(df.query('n_partial >= @n_thresh').query('r_partial > 0').groupby('site_comparison').size())
````

````python
oldshared = df.query('n_partial >= @n_thresh').query('r_site12 > 0').query('q_site12 < 0.1')
oldshared.shape
````

````python
sns.set_style('white')
fig, ax = plt.subplots(1, 3, figsize=(12,4))
i = 0
comps = ['bal-throat_swab', 'bal-gastric_fluid', 'gastric_fluid-throat_swab']
for s in comps:
    subdf = oldshared.query('site_comparison == @s')
    ax[i].scatter(subdf['r_site12'], subdf['r_partial'])
    ax[i].plot([-0.1, 1], [-0.1, 1], '--')
    ax[i].plot([-0.1, 1], [0, 0], '--')
    ax[i].set_xlim([-0.1, 1])
    ax[i].set_ylim([-0.1, 1])
    ax[i].set_xlabel('normal corr')
    if i == 0:
        ax[i].set_ylabel('partial corr')
    ax[i].set_title(s)
    i += 1
    
````

````python
newshared = df.query('n_partial >= @n_thresh').query('r_partial > 0').query('q_partial < 0.1')
newshared.shape
````

````python
sns.set_style('white')
fig, ax = plt.subplots(1, 3, figsize=(12,4))
i = 0
comps = ['bal-throat_swab', 'bal-gastric_fluid', 'gastric_fluid-throat_swab']
for s in comps:
    subdf = newshared.query('site_comparison == @s')
    ax[i].scatter(subdf['r_site12'], subdf['r_partial'])
    ax[i].plot([-0.1, 1], [-0.1, 1], '--')
    ax[i].plot([-0.1, 1], [0, 0], '--')
    ax[i].set_xlim([-0.1, 1])
    ax[i].set_ylim([-0.1, 1])
    ax[i].set_xlabel('normal corr')
    if i == 0:
        ax[i].set_ylabel('partial corr')
    ax[i].set_title(s)
    i += 1
    
````

One more test: how do the OTUs picked out by the two methods compare?

````python
df.loc[((df['q_site12'] < 0.1) & (df['r_site12'] > 0)), 'shared_old'] = 1
df.loc[((df['q_partial'] < 0.1) & (df['r_partial'] > 0)), 'shared_partial'] = 1
````

````python
compare_shared = df\
    .dropna(subset=['shared_old', 'shared_partial'], how='all')\
    [['otu', 'site_comparison', 'shared_old', 'shared_partial']]
````

````python
fig, ax = plt.subplots(3, 1, figsize=(4, 16))
i = 0
for g in comps:
    subdf = compare_shared\
                .query('site_comparison == @g')\
                .sort_values(by=['shared_partial', 'shared_old'])\
                [['otu', 'shared_partial', 'shared_old']]\
                .fillna(0)
    ax[i].imshow(subdf[['shared_partial', 'shared_old']], 
                 interpolation='nearest', aspect='auto')    
    ax[i].yaxis.tick_right()
    ax[i].yaxis.set_label_position('right')
    ax[i].set_yticks(np.arange(0,len(subdf['otu'])))
    ax[i].set_yticklabels(subdf['otu'], fontsize='small', va='bottom')
    ax[i].set_title(g)
    ax[i].set_xticks([0, 1])
    ax[i].set_xticklabels(['partial', 'old'])

    i += 1
````

````python
g = 'bal-throat_swab'
subdf = compare_shared\
            .query('site_comparison == @g')\
            .sort_values(by=['shared_partial', 'shared_old'])\
            [['otu', 'shared_partial', 'shared_old']]\
            .fillna(0)
subdf.tail()
````

### For group meeting, make a bar plot of shared OTUs per phylum per site.

````python
compare_shared.head()
````

````python
phylocols = ['kingdom', 'phylum', 'class', 'order', 'family', 'genus', 'species', 'denovo']
compare_shared[phylocols] = compare_shared['otu'].str.split(';', expand=True)
compare_shared.head()
````

````python
shared_counts = compare_shared\
    .dropna(subset=['shared_partial'])\
    .groupby(['site_comparison', 'phylum'])\
    .count()\
    ['shared_partial']\
    .reset_index()
shared_counts['phylum'] = shared_counts['phylum']\
    .replace('p__', 'p__Unknown')\
    .replace('p__Candidatus_Saccharibacteria', 'p__Saccharibacteria')\
    .replace('p__Cyanobacteria/Chloroplast', 'p__Cyanobacteria/\nChloroplast')\
    .str[3:]
shared_counts.head()
````

````python
fig, ax = plt.subplots(figsize=(12,3))
sns.barplot(data=shared_counts, x='phylum', y='shared_partial', hue='site_comparison',
           hue_order=comps, ax=ax)
ax.legend(loc='upper right')
ax.set_ylabel('Number OTUs exchange')
ax.set_xlabel('')
````

````python
shared_counts.groupby('site_comparison').sum()
````
